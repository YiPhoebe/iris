```python
# 1-1. 데이터 불러오기 및 구조 확인 (10점)

# - **요구사항:**
#     1. `sklearn.datasets.load_iris()` 또는 UCI Iris CSV 파일을 이용하여 데이터를 불러오고, 이를 pandas DataFrame으로 변환하시오.
#     2. 데이터의 크기(행, 열)와 각 변수의 데이터 타입을 출력하는 코드를 작성하시오.
#     3. `describe()` 함수를 사용하여 각 변수(SePalLength, SepalWidth, PetalLength, PetalWidth)의 최소값, 평균, 표준편차 등의 기초 통계량을 산출하고,
#     주요 통계 수치에 대한 간단한 해석을 첨부하시오.

# 필요한 라이브러리
from sklearn.datasets import load_iris  # scikit-learn의 샘플 데이터 중 iris를 로드 하기 위해 임포트
import pandas as pd     # 데이터 프레임으로 변환을 위해 임포트
import numpy as np      # 수학 연산 위해 임포트
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# info()로 결측치 및 데이터 타입 확인
iris = sns.load_dataset('iris')
iris.info()

# 데이터 불러오기
iris = load_iris()

# 데이터프레임으로 변환. df 는 데이터 프레임
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df.describe()


```
