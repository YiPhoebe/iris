2. 시각화 결과를 토대로 관찰할 수 있는 변수 간의 관계나 특이한 패턴(예: 군집, 분포의 비대칭성 등)을 서술하시오.

```python
# ** gpt의 도움..

# 품종 데이터 추가 (iris.target의 값이 0, 1, 2이므로 이를 문자열로 매핑)
df["species"] = iris["target"]
df["species"] = df["species"].map({0: "setosa", 1: "versicolor", 2: "virginica"})

# Pairplot을 사용하여 변수 간 산점도 및 히스토그램 시각화 (품종별 색상 구분)
sns.pairplot(df, hue="species", diag_kind="hist", markers=["o", "s", "D"])

# 그래프 제목 추가
plt.suptitle("Pairplot of Iris Dataset with Species", y=1.02)

# 그래프 출력
plt.show()
```

```python

# 각 품종(Setosa, Versicolor, Virginica)별로 Boxplot을 작성하여 꽃받침과 꽃잎 변수들의 분포를 비교하시오.
sns.set_style("whitegrid")
plt.figure(figsize=(8,6))

sns.boxplot(data=df)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="sepal width (cm)", data=df)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="sepal length (cm)", data=df)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="petal width (cm)", data=df)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="petal length (cm)", data=df)
plt.show()
```

sepal width에 이상치 존재.

```python
# gpt 도움..
# 박스플롯 그리기 - 품종별 꽃받침 & 꽃잎 비교
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Sepal Length Boxplot (꽃받침 길이)
sns.boxplot(x="species", y="sepal length (cm)", data=df, ax=axes[0, 0])
axes[0, 0].set_title("Sepal Length by Species")

# Sepal Width Boxplot (꽃받침 너비)
sns.boxplot(x="species", y="sepal width (cm)", data=df, ax=axes[0, 1])
axes[0, 1].set_title("Sepal Width by Species")

# Petal Length Boxplot (꽃잎 길이)
sns.boxplot(x="species", y="petal length (cm)", data=df, ax=axes[1, 0])
axes[1, 0].set_title("Petal Length by Species")

# Petal Width Boxplot (꽃잎 너비)
sns.boxplot(x="species", y="petal width (cm)", data=df, ax=axes[1, 1])
axes[1, 1].set_title("Petal Width by Species")

# 그래프 정리 및 출력
plt.tight_layout()
plt.show()
```
